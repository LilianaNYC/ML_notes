<ul>
    <li>SVM is a linear classifier that can be viewed as an extension of the Perceptron. The Perceptron 
        guaranteed that you find <b>aM</b> hyperplane if it exists. The SVM finds the maximum margin separating 
        hyperplane.</li>
    <li><b>Kernel</b> computes relationships between the observations in a higher dimension then find a support 
        vector classifier based on the high dimensional relationship</li>
    <ul>
        <li>linear kernel</li>
        <li>polynomial kernel</li>
        <li>RBF kernel</li>
    </ul>
    <li>uses the concept of margins to come up with parameters</li>
    <li>can add a regularization parameter to the loss function to maximize the margin</li>
    <li>Question: What is the best separating hyperplane? SVM Answer: The one that maximizes the distance to the 
        closest data points from both classes. We say it is the hyperplane with maximum margin.</li>

<h1>Perceptron</h1>
<ul>
    <li>At its core, a perceptron consists of input nodes, weights associated with each input, a summation 
        function, an activation function, and an output. The input nodes receive numerical inputs, which are 
        multiplied by corresponding weights. These weighted inputs are then summed, and the resulting sum is 
        passed through an activation function. The activation function introduces non-linearity and determines 
        whether the perceptron outputs a 0 or a 1, based on a predefined threshold.</li>
    <li><b>Assumptions</b>: the data must be linear separable</li>
    <li><a href="https://www.cs.cornell.edu/courses/cs4780/2018fa/lectures/lecturenote03.html">resource</a></li>
</ul>

<b>Steps</b>
<ol>
    <li>Initialize with random weights and bias 
        w<sub>0</sub>+w<sub>1</sub>x<sub>1</sub>+w<sub>2</sub>x<sub>2</sub></li>
    <li>Each feature is multiplied by its weight and the results are summed up with the bias term. Then passes 
        thru an activation function to produce the prediction</li>
    <li>If predictions are correct then no changes are made to the weights else the weights are adjusted to 
        reduce the error</li>
    <ul>
        <li>the adjustment is based on a learning rate and the difference between the predicted and actual 
            values</li></ul>
    <li>repeat steps 2-3 until the algorith converges (max iterations are reached or weights no longer change 
            significantly</li>
