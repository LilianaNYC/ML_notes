<ul>
    <li>Data augmentation:
        <ul>
            <li>artifically expand the size of training set by creating modified data from the existing one</li>
            <li>prevents overfitting</li>
            <li>typically used for learning mmodels</li>
        </ul>
    <li><b>Neural Network: </b>are composed of a collection of neurons that are organized in layers, each with 
        their own learnings weights and biases.</li>
    <li>CNN-classifier: it is a neural network algorithm used to recognize patterns in data</li>
    <li>Convolution layers: uses information from adjacent pixel to down sample the imagine into features by 
        convolution and then uses prediction layers to predict the target values (use filters and kernels). It 
        also uses back propagation to extract different features of the image</li>
    <li>Max pooling layer: helps reduce the spatial size of the convolved features and reduce overfitting by 
        providing an abstracted representation of them. It is similar to the convolution layer but instead of 
        taking the dot product between the input and the kernel we take the max of the region from the input 
        overlapped by the kernel.</li>
    <li>Activation Functions: </li>
        <ul>
            <li>RELU: introduce non-linearity to the model which allows it to learn complex functional 
                mappings between the inputs and response variables</li>
            <li>SIGMOID</li>
            <li>TANH</li>
    </ul>
    <li>Fully connected layers: the input layers nodes are connected to every node in the second layer. Adding a 
        fully connected layers helps learn non-linear combinations of high level features outputed by the 
        convolutional layers (activation functions and dropout layers are used between two consecutive fully 
        connected layers to introduce a non linearity and reduce overfit)</li>
