<h1>K-means</h1>

<ul>
    <li><b>Notes</b>
        <ul>
            <li>To find the most optimal cluster run the model more than 50 times and pick the set of cluster 
                that gives the lower the cost(j)</li>
            <li><b>Lost Function:</b>Loss = ∑<sub>i=1</sub><sup>N</sup> ∑<sub>j=1</sub><sup>K</sup> 
                ||x<sub>i</sub> - μ<sub>j</sub>||<sup>2</sup> </li>
            <li>K-means inertia measures how well a dataset was clustered by K-mmeans (measures the distance 
                between each data point and its centroid, squaring this distance and summing these squares 
                accross one cluster)</li>
            <li>Uses euclidian distance</li>
            <li>Objective function is non-convex (can only find local minimum) and each run the results are 
                different</li>
            <li>It is an iterative algorithm</li>
            <li>Uses coordinate descent (updates one parameter first then the other)</li>
        </ul>
    <li><b>Steps</b></li>
        <ol>
            <li>Select the number of clusters you want to identify</li>
            <li>Select the centroids randomly</li>
            <li>Measure the point between 1st and kth centroids</li>
            <li>Assign each point to its closest centroid (using minimum distance)</li>
            <li>Recompute the ceontroids by taking the average of the locations</li>
            <li>Repeat</li>
        </ol>
</ul>
