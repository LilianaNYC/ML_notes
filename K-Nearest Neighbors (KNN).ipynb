<h1>Notes</h1>
<ul>
    <li>It's a supervise ML algorithm mostly used for classification but it can be used regression as well</li>
    <li>It can be used as unsupervise by using the KNN (euclidian/etc.) distance</li>
    <li>It is based on feature similarity</li> 
    <li>To choose a value of K then compute the &radic;N where N=total_number_of_points (works better when it's an odd number)</li>
    <li>Data needs to be normalized/scaled/standarize</li>
    <li>When K>N each test instance will be based on the majority class or average value of the entire dataset. This can result in a uniform or constant prediction for all test instances. This is something to consider when using cross-validation</li>
    <li>Multicollinearity is not an issue with K-NN</li>
</ul>


<h1>Bias-Variance</h1>
<ul>
    <li>A low value of K tends to result in higher variance and lower bias</li>
    <ul>
        <li>A lower value of K leads to a more complex decision boundary that closely fits the training data</li>
        <li>A lower value of K means that the predictions are influenced by fewer neighbors, resulting in higher sensitivity to the specific training instances</li>
    </ul>
    <li>Increasing the value of K in KNN tends to increase the bias and decrease the variance of the model</li>
    <ul>
        <li>This means that the model becomes more biased towards the majority class or prevalent patterns in the data</li>
        <li>performs better when there is more outliers and noise
        <li>However, it may sacrifice the ability to capture finer details or local patterns, potentially resulting in a loss of predictive accuracy</li>
    </ul>
</ul>




<h1>Extreme Cases</h1>
<ul>
    <li>If <b>K=N</b> 
        <ul>
            <li>All data points in the training dataset are treated as neighbors for each test instance</li>
            <li>The algorithm calculates the distances between the test instance and all training instances which can be computationally expensive for large datasets</li>
            <li>It results in lower variance and higher bias compared to smaller values of K </li>
            <li>By considering all data points, the model becomes less flexible and more biased towards the overall distribution of the data. However, it may provide better generalization and be less prone to overfitting</li>
        </ul>
    <li>If <b>K=1</b></li>
    <li>Can lead to overfitting, especially when the training data contains noise or outliers</li>
    <li>The model becomes highly sensitive to the training instances and may not generalize well to unseen data</li>
    <li>It can result in high variance, meaning the predictions may vary greatly depending on the specific training instances</li>

<h1>Assumptions</h1>
<ol>
    <li>Feature relevance: Features that are irrelevant or noisy can negatively impact the algorithm's performance</li>
    <li>Distance metric: It assumes that the chosen distance metric (such as Euclidean distance or Manhattan distance) accurately represents the dissimilarity between instances.</li>
    <li>Locality assumption: It implies that data points in close proximity are likely to share similar characteristics and belong to the same class</li>
    <li>Data distribution: The distribution of data is locally smooth, meaning that instances of the same class tend to be clustered together in the feature space </li>
    <li>Training data representativeness: The training dataset is representative of the underlying population or data distribution. It assumes that the training instances adequately cover the range of possible input variations and reflect the true relationships between features and target.
    

<h1>Algo Steps</h1>
<ol>
    <li>Choose the value of K</li>
    <li>Calculate distances: For each instance in the test dataset, calculate the distance between that instance and all instances in the training dataset</li>
    <li>Find the nearest neighbors: Select the K instances from the training dataset that are closest to the instance being predicted, based on the calculated distances</li>
    <li>Make predictions: Based on the class labels of the K nearest neighbors, determine the class or target value for the instance being predicted. In classification tasks, the majority class among the K nearest neighbors is typically assigned as the predicted class for the test instance. In regression tasks, the average or weighted average of the target values of the nearest neighbors is often used as the predicted value</li>
    <li>Evaluate the model</li>
    <li>Iterate and optimize: Depending on the performance of the KNN classifier, you can iterate and optimize the model by adjusting parameters such as K, feature selection, or data preprocessing techniques</li>
 
</ol>
